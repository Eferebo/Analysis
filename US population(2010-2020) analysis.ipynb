```python
# Importing modulus
import pandas as pd
import seaborn as sns
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading the data into a dataframe
data = pd.read_csv("/Users/HP/Downloads/US_States_Populations.csv")
```

```python
data.head()
```

```python
max_pop_2020 = data.groupby("State")["Pop in 2020"].max()
max_pop_2020.head().sort_values()
```

```python
max_pop_2010 = data.groupby("State")["Pop in 2010"].max()
```

```python

print('Minimum  number of population in 2020 is',max_pop_2020.min())
```

```python

print(' Maximum number of population in 2020 is',max_pop_2020.max())
```

```python

sum_2020 = max_pop_2020.sum()
print('Total number of population in USA in 2020 is',sum_2020)
```

```python

sum_2010=max_pop_2010.sum()
print('Total number of population in USA in 2010 is',sum_2010)
```

## What is the increase in population from 2010 to 2020

```python
sum_2020 - sum_2010
```

#### The U.S population from 2010 to 2020 was increased by 22,226,684

```python

print(' Average number of population in USA in 2020 is',max_pop_2020.mean())
```

```python

print('Average number of population in USA in 2010 is',max_pop_2010.mean())
```

```python
plt.rcParams['figure.figsize'] = (30, 20)
ax = sns.barplot(data,x="Pop in 2020",y='State')
plt.xlabel('Population rate')
plt.ylabel('State')
plt.title("U.S Population of each States in 2020");
```

```python
plt.rcParams['figure.figsize'] = (30, 20)
ax = sns.barplot(data,x="Pop in 2010",y='State')
plt.xlabel('Population rate')
plt.ylabel('State')
plt.title("U.S Population of each States in 2010");
```

```python
plt.rcParams['figure.figsize'] = (25, 11)
ax = sns.barplot(data,x="Growth",y='State',color='blue')
plt.xlabel('Growth')
plt.ylabel('State')
plt.title("Growth in population of States in the U.S");
```

### The above vizualization shows the increase (to the right) and the decreace (to the left) of the population in States of the U.S from 2010 to 2020
